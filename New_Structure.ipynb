```python
author_demos = pd.read_csv('complete_author_demographics.csv')
author_dict = author_demos.set_index('name').T.to_dict('list')
pub_sample = pd.read_csv('Engineering/Publications_Simple_College of Engineering_20190403.csv')
col_name = 'College of Engineering'
```

```python
pub_sample = pub_sample[['Name','Username','Email','Primary group descriptor', 'Authors']]
```

```python
def get_stat(a, b, author_stats):   
    if a in author_stats:
        stat_a = author_stats[a]
    else:
        stat_a = [0,0,0] #,set(),set(),set()
    if b in author_stats:
        stat_b = author_stats[b]
    else:
        stat_b = [0,0,0] 
    return stat_a, stat_b
```

```python
author_stats = {}
for index, row in pub_sample.iterrows():
    #print(index)
    try:
        nameArr = row['Name'].split(',')
        name = nameArr[0].lower() + nameArr[1][0:2].lower()
        author_sep = row['Authors'].split(', ') # , and space
    except:
        print(str(index)+" "+row['Name'])
        continue
    #author_sep = author_sep.split(' ')
    #author_list = author_sep[0].lower() + author_sep[1].lower()
    #print(author_sep[1].split(' ')[1][0:1].lower())
    co_name = []
    for co_a in author_sep:
        
        #print(co_name)
        try:
            co_name.append(co_a.split(' ')[0].lower()+" "+co_a.split(' ')[1][0:1].lower()) #Last name + space + first name first char
       # print(co_name)
        except:
            continue
    
    
    for i in range(len(co_name)):
        for j in range(i + 1, len(co_name)):
            a = co_name[i]
            b = co_name[j]
            if a in author_dict and b in author_dict:
                stat_a, stat_b = get_stat(a, b, author_stats)
                #print(stat_a)
                #Get the dept and col names
                author_a_dept = author_dict[a][7].lower()
                author_b_dept = author_dict[b][7].lower()
                #print(author_a_dept)
                #print(author_b_dept)
                try:
                    author_a_col = author_dict[a][8].lower()
                    author_b_col = author_dict[b][8].lower()
                except:
                    pass
                #print(author_a_col)
                #print(author_b_col)
                # Foramt stat - list{dept_count, col_count, uni_count}
                if author_a_dept == author_b_dept:
                    stat_a[0] += 1
                    stat_b[0] += 1
                    #stat_a[3].add(b)
                    #stat_b[3].add(a)
                elif author_a_col == col_name.lower() and author_b_col == col_name.lower():
                    #print("Yes")
                    stat_a[1] += 1
                    stat_b[1] += 1
                    #stat_a[4].add(b)
                    #stat_b[4].add(a)
                else:
                    stat_a[2] += 1
                    stat_b[2] += 1

                author_stats[a] = stat_a
                author_stats[b] = stat_b
                #print(author_stats)
```

```python
author_statistics = pd.DataFrame.from_dict(author_stats, orient='index', columns=['dept_count', 'college_count', 'uni_count'])
author_statistics['name'] = author_statistics.index
author_statistics= author_statistics.reset_index()
author_statistics = pd.merge(author_statistics, author_demos, how="left", on="name")
author_statistics = author_statistics.drop(['Unnamed: 0','Unnamed: 0.1','index'],  axis=1)
author_statistics = author_statistics[['netid', 'name', 'lastname', 'firstname', 'middle_name','title','dept_name','college_name\
','dept_count','college_count','uni_count']]
```
